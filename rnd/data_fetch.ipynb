### Get the data

```python
import httpx
from pprint import pprint
```

```python
github_url = "https://raw.githubusercontent.com/yuhonas/free-exercise-db/main/dist/exercises.json"

with httpx.Client() as client:
    resp = client.get(github_url)
    resp.raise_for_status()
    data = resp.json()
```

```python
pprint(data[0])
```
